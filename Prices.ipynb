```
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
```

```
housing_data = pd.read_csv("sample_data/california_housing_train.csv")
```

```
housing_data.head()
```

```
housing_data.describe()
```

```
target = housing_data.median_house_value
features = ["total_rooms", "total_bedrooms", "housing_median_age", "households"]
input_data = housing_data[features]
```

```
input_data.head()
```

```
housing_price_model = DecisionTreeRegressor(max_leaf_nodes=10000, random_state=1)
housing_price_model.fit(input_data, target)
```

```
predictions = housing_price_model.predict(input_data.head())
print(predictions)
```

PREDICT NEW AREA

```
room_new = int(input("Enter number of rooms in the new area: "))
bedroom_new = int(input("Enter the number of bedrooms in the new area: "))
median_age_new = int(input("Enter the median age of housing in the new area in years: "))
households_new = int(input("Enter the number of households in the new area: "))
new_area={"total_rooms":[room_new],
          "total_bedrooms":[bedroom_new],
          "housing_median_age":[median_age_new],
          "households":[households_new]}

new_area_df = pd.DataFrame.from_dict(new_area)

predictions = housing_price_model.predict(new_area_df.head())
print(predictions)
```

```
predictions = housing_price_model.predict(input_data)
mean_absolute_error(target, predictions)
```

```
test_data = pd.read_csv('sample_data/california_housing_test.csv')
target = test_data.median_house_value
input_data = test_data[features]
predictions = housing_price_model.predict(input_data)
mean_absolute_error(target, predictions)
```
